# Problem 1: Geweke Convergence Test

In the lecture we have seen the Geweke test as one option to test for convergence of our metropolis hastings chain. Describe in your own words how the Geweke test works and its limitations. 

We provide you with the following code for the Geweke test. Write comments for the marked lines and any lines you think would be good to explain in addition. 

Write a short explanation of what the `rhot` function does and why it is necessary. 

###  Answer
**1. Geweke Test:**
It is to test whether the mean estimates have converged by comparing means from the early and latter part of the Markov chain. Normally, in a long enough chain whose trace plot suggests convergence, we assume the second half of the chain has converged to the target distribution. In our example here, we just treat the first 10% as burn in(not converged) part and throw it away. And we then divide the trace into different windows. We used z test to test whether the mean of each window in the first 50% is significantly different from the last 50%. If not, we conclude that the target distribution already converged somewhere in the early window of the chain we test. 

**2. Limitation of Geweke Test:**  1). It is sensitive to the specification of the spectral window. 2). Not specify a procedure for applying the diagnostic but leave to the subjective choice of the users.

**3. rhot function:** The rhot function here calculates the autocorrelation. Because the naive standard deviation is the standard error of the mean which captures simulation error of the mean rather than posterior uncertainty. The time-series standard error adjusts the “na¨ıve” standard error for autocorrelation.

```python
# Calculate the correlation between the early and later part of x at break t.
def rhot(x, t):
    n = len(x)
    return np.corrcoef(x[0:(n-t)], x[t:n])[0,1]

# Function to do the Geweke test.
def Geweke(trace, intervals, length):
    # the size of each window
    nsl=length
    # caculate the step size 
    jump = int(0.9*len(trace)/(2*intervals))
    # the starting point (discard the first 10%)
    first = 0.1*len(trace)
    # create a empty array to store the z value of each test
    z =np.empty(intervals)
    # we divide the trace into 2*intervals sub parts and pair these parts.
    # the for loop loop through each pair and do the z test to compare the mean of the two subparts in the pair.
    for k in np.arange(0, intervals):
        # calculate the starting point of the early subpart in the pair
        baga = np.int(first+k*jump)
        # calculate the starting point of the latter subpart in the pair
        bagb = len(trace)/2 + k*jump
        
        # use the starting point and the length to extract the two subparts out
        sub_trace_a = trace[baga:baga+nsl]
        sub_trace_b = trace[bagb:bagb+nsl]
        
        # calcualte the mean of each subpart
        theta_a = np.mean(sub_trace_a)
        theta_b = np.mean(sub_trace_b)
        rho_a, rho_b = 1.0, 1.0
        # 
        for i in xrange(int(0.1*nsl)):
            # divide each window into 10 parts 
            # calculate 2 times the sum of the correlation between each part and the part after it 
            # to get the autocorrelation for the variance
            rho_a += 2*rhot(sub_trace_a, i+1)
            rho_b += 2*rhot(sub_trace_b, i+1)
            
        # use the autocorrelation we calculated to adjust the naive variance
        var_a  = np.var(sub_trace_a)*rho_a/length
        var_b  = np.var(sub_trace_b)*rho_b/length
        
        # calculate the z value
        z[k] = (theta_a-theta_b)/np.sqrt( var_a + var_b)
    
    return z
```

### 2.2 Derive and compute the maximum likelihood solution for a Poisson distribution.

If $x_{1}, x_{2}, \dots, x_{n}$ follows the Poisson distribution

$$ P(x_i|\lambda) = \frac{\lambda^{x_i}}{x_i!}e^{-\lambda}, \lambda>0$$

Likelihood function:
$$L(\lambda)=p(x_1,x_2,...,x_n|\lambda)=\frac{\lambda^{\sum_ix_i}e^{-n\lambda}}{\prod_{i=1}^{n}x_i!}$$
$$L(\lambda)\propto \lambda^{\sum_i x_i}e^{-n\lambda}$$

Log likelihood:
$$l(\lambda) = C + [\ln(\lambda)\sum_i x_i] -n\lambda$$

Take the derivative of the Log likelihood function:
$$\frac{dl(\lambda)}{d\lambda}=\frac{\sum_i^n x_i}{\lambda}-n=0$$ 
$$\rightarrow \hat{\lambda}_{mle} = \frac{\sum_i^n x_i}{n}$$

### 2.3 Compare this to the Bayesian solution with a prior of your choice and using your own implementation of Metropolis Hastings to sample from the posterior. Make sure to describe why you chose this prior, as well as the specifics of your Metropolis Hastings implementation.

I will choose to use a gamma distribution as my prior. 
* Gamma distribution and Poisson distribution have the same support: 0 to infinity.
* In Bayesian statistics, the gamma distribution is used as a conjugate prior distribution for various types of inverse scale (aka rate) parameters, such as the λ of an exponential distribution or a Poisson distribution.

Likelihood function:
$$L(\lambda)=p(x_1,x_2,...,x_n|\lambda)=\frac{\lambda^{\sum_ix_i}e^{-n\lambda}}{\prod_{i=1}^{n}x_i!}, \lambda>0$$

Gamma Prior:
$$p(\lambda) = \frac{\beta^{\alpha}}{\Gamma(\alpha)}\lambda^{\alpha-1}e^{-\beta\lambda}$$

Posterior:
$$p(\lambda|x_1,x_2,...,x_n)\propto p(x_1,x_2,...,x_n|\lambda)* p(\lambda) = \frac{\beta^{\alpha}}{\prod_{i=1}^nx_i!\Gamma(\alpha)}\lambda^{(\alpha-1)+\sum_ix_i}e^{-\lambda(\beta+n)}$$
$$\propto \lambda^{(\alpha-1)+\sum_ix_i}e^{-\lambda(\beta+n)}$$

Take the log of the posterior:
$$\ln(p(\lambda|x_1,x_2,...,x_n)) = C+(\alpha-1+\sum_ix_i)\ln(\lambda)-\lambda(\beta+n)$$

Take the derivative of the log of the posterior:
$$\frac{d\ln p(\lambda|x_1,x_2,...,x_n)}{d\lambda}= \frac{\alpha-1+\sum_ix_i}{\lambda}-(\beta+n)$$
$$\hat{\lambda}_{map} = \frac{\alpha-1+\sum_ix_i}{\beta+n}$$

When n get larger and larger, our map estimator get closer and closer to our mle estimate.

### 2.3 Using Metropolis Hastings to sample from the posterior.
In our problem, we do not know much information about the response time so I set a vague prior which is Gamma($\alpha=1, \beta=0.1$)

The posterior is also a gamma distribution with parameters $\alpha＝1+\sum_i x_i$, $\beta=0.1+n$. 


```python
# Set paramter for posterior function
alpha=1+sum(data.values)[0]
beta=0.1+len(data.values)

# Caculate posterior
calc_posterior = lambda x,alpha=alpha,beta=beta: scipy.stats.gamma.pdf(x, alpha, scale = 1.0/beta)
```

```python
def metropolis(n_iterations, initial_values, prop_std= 1, doPrint=True):
    #################################################################
    # function to sample using Metropolis 
    # (assumes proposal distribution is symmetric)
    # There is only one parameter here to estimate 
    #
    # n_iterations: number of iterations
    # initial_values: start position for our chain
    # prop_std: standard deviation for Gaussian proposal distribution 
    ##################################################################
    
    # Initialize trace for parameters
    trace = np.empty((n_iterations+1))
    
    # Set initial values
    trace[0] = initial_values
        
    # Calculate joint posterior for initial values
    # the * assigns the arguments of the function according to the list elements
    current_prob = calc_posterior(trace[0])
    
    # Initialize acceptance counts
    # We can use this to tune our step size
    accepted = 0
    
    for i in range(n_iterations):
    
        if not i%10000 and doPrint: 
            print 'Iterations left: ', n_iterations-i
    
        # Grab current parameter values
        current_params = trace[i]
        
        # proposed new value
        theta = np.random.normal(current_params, prop_std)    
        p = theta
    
        # Calculate posterior with proposed value
        proposed_prob = calc_posterior(p)
    
        # Log-acceptance rate
        a = proposed_prob/current_prob
    
        # prob of acceptance
        prob = min(1, a)
        # Sample a uniform random variate
        u = np.random.uniform()

        # Test proposed value
        if (u) < a:
                # Accept
            trace[i+1] = theta
            current_prob = proposed_prob
            accepted += 1
        else:
            # Stay put
            trace[i+1] = trace[i]
    
    # return our samples and the number of accepted steps
    return trace, accepted
```

### Answer
This markov chain is not irreducible but it is aperiodic.

**Irreducible:**  A Markov chain is irreducible if there is a non-zero probability of transitioning (even if in more than one step) from any state to any other state. This chain is not irreducible because we can not go from state 1,2 or 3 to state 4 or 5. And we can not go from state 4 or 5 to state 1,2 or 3 either.

**Aperiodic:** A state i has period k if any return to state i must occur in multiples of k time steps. If k = 1, then the state is said to be aperiodic: returns to state i can occur at irregular times. This markov chain is aperiodic because none state in our chain has a period k>1.

**1->1:** 1->2->1, 1->3->1, 1->2->3->1, 1->2->3->2->1 ... gcd(3,4,5,...) =1

**2->2:** 2->1->2, 2->1->3->2, 2->1->3->1->2, ... gcd(3,4,5,...)=1

# Problem 4: The Evidence 

In lectures we mostly concentrate on the likelihood and the prior and regard the evidence as a mere normalization factor. However, the evidence can be quite useful. In this problem you will compare different models by computing the evidence for each model, aka the probability that randomly selected parameters from a given model class would generate the data $X$.

As our models we compare polynomials of degree 0 to 4. For example for degree 2 we have $y = a_0 + a_1 \cdot x + a_2 \cdot x^2 + \epsilon$ where $\epsilon \sim N(0, \sigma) $ and $\theta = [a_0, a_1, a_2, \sigma]$. Assume that for all polynomials $a_i \sim \text{Expo}(1)$ and $\sigma \sim \text{Inverse Gamma}(1,1)$.


Bayes' theorem states that:
$$
P(\theta|x)=\frac{P(x|\theta)P(\theta)}{P(x)}
$$
$P(\theta)$ is the prior, $P(x|\theta)$ is the likelihood, $P(x)$ is called the evidence, and $P(\theta|x)$ is the posterior.

Your tasks are:

* Write down the mathematical equation of the evidence in terms of the likelihood and the prior. You don't need to solve the integral analytically, just write down its formula.
* Now you have the evidence in the form of an integral. Solve it by using importance sampling. What is a good choice for your importance sampling distribution?
* Compare the evidence for polynomials of degree 0 to 4. Which polynomial wins?

### Answer:
**Mathematical Equation of the Evidence:** $P(x)=\int P(\theta)P(X|\theta)d\theta$

**Choice for importance sampling distribution:** 
$\int P(\theta)P(X|\theta)d\theta=\int P(X|\theta)P(\theta)d\theta\approx \frac{1}{N}\sum P(x|\theta)$ where $\theta$ are generated from P($\theta$) So the choice would just be the distribution of $\theta$ where $\theta = [a_i, \sigma]$. 

Assume that for all polynomials $a_i \sim \text{Expo}(1)$ and $\sigma \sim \text{Inverse Gamma}(1,1)$..

Our P(x|$\theta$) here is the pdf of y|$\theta$. If we have $y = a_0 + a_1 \cdot x + a_2 \cdot x^2 + \epsilon$ where $\epsilon \sim N(0, \sigma)$. 

y|$\theta$ ~ N($a_0 + a_1 \cdot x + a_2 \cdot x^2, \sigma$)

```python
# Here is your data for the problem:
data = np.array([[ -1.85519254,  -2.7009541 ],
       [  4.38291824,  19.61735369],
       [  2.29495208,   3.96481822],
       [  0.02075668,   8.00646088],
       [  0.54097177,   2.8872262 ]])

x=data[:,0]
y=data[:,1]
```

```python
# Number of draws 
N = 1000

# ============================================
# IMPORTANCE SAMPLING 
# ============================================
# function to calculate product
import operator
def prod(factors):
    return reduce(operator.mul, factors, 1)

def IS(N,f,x,y,degree):
    Iis = np.zeros(N)
    # draw sigma from Inverse Gamma(1,1)
    sig_list = scipy.stats.invgamma.rvs(a=1., scale=1., size=N).tolist()
    for k in np.arange(0,N):
        # draw ai from exponential distribution 
        a = scipy.stats.expon.rvs(1, size=(degree+1))
        # get sigma for this iteration
        sigma = sig_list[k]
        # calculate f(y|theta)
        p = [None]*len(y)
        for i in range(len(y)):
            p[i] = scipy.stats.norm(f(a,x[i]), sigma).pdf(y[i])
        # computing our estimates
        Iis[k] = prod(p)
    return Iis
```

```python
print "Degree 0 Evidence MC estimate: ", np.mean(Iis0)
print "Degree 1 Evidence MC estimate: ", np.mean(Iis1)
print "Degree 2 Evidence MC estimate: ", np.mean(Iis2)
print "Degree 3 Evidence MC estimate: ", np.mean(Iis3)
print "Degree 4 Evidence MC estimate: ", np.mean(Iis4)
```

Degree 1 has the largest evidence so degree 1 wins.

# Problem 5: Which YouTube Videos to Watch

Youtube videos have a like and an unlike flag. We can use these up and down votes on the videos to determine if a video is worth watching. However, it is not immediately obvious how to rank a video with just 3 up and 0 down rankings against a video with 300 up and 100 down votes. We will address this problem using a Bayesian approach.

Build two Bayesian models for the average upvote rate of a video. Both models should use the same likelihood, but different priors. Use one prior where people in general are rather undecided about videos, and one where people tend to be very opinionated. Compare the resulting posteriors for each video. How does the different choice of prior change your results? 

Given that there are so many videos on YouTube we want a really conservative way to decide if it is worth watching. Compute the 5th percentile for the posterior of each video and rank the videos according to this value. Is the 5th percentile a good indicator for the ranking? What are the benefits, what are potential drawbacks?

```python
# Here is the [upvote, downvote] data for 4 different videos:
video_votes = np.array([[3,0],[300,100],[2,2],[200,100]])
```

## Choose prior and likelihood
I choose to use binomial as the likelihood and each video has their own parameter for the binomial distribution.In this model the up rate $\theta$ as a binomial. This gives the Likelihood Function:

$$P(D|\theta) = \frac{n!}{(n-k)!k!}\theta^{k}(1-\theta)^{n-k}$$

And I choose to use beta distribution for the prior, which is the same for all videos. However, there are 2 scenarios.

Ones is people in general are rather undecided about videos. In this case, I set the prior as Beta(2,2).

Another is people in general are very opinionated about videos. In this case, I set the prior as Beta(0.5,0.5).

we calculate the posterior:

$$P(\theta|D) \propto P(D|\theta)P(\theta) = Beta(\theta,\alpha+number of successes,\beta +number of failures)$$


```python
# undecided prior
N = 10000

# set shape parameters
betaParamsExample = [2,2]

# sample from the distributions N times
betaDistExample = np.random.beta(betaParamsExample[0],betaParamsExample[1],N)

plt.hist(betaDistExample,25,normed=1)
plt.xlabel("Upvote Rate")
plt.ylabel("Probability Density")
plt.title("PDF of Undecided Prior Distribution")
plt.xlim(0,1)
```

```python
# opinionated prior
N = 10000

# set shape parameters
betaParamsExample = [0.5,0.5]

# sample from the distributions N times
betaDistExample = np.random.beta(betaParamsExample[0],betaParamsExample[1],N)

plt.hist(betaDistExample,25,normed=1)
plt.xlabel("Upvote Rate")
plt.ylabel("Probability Density")
plt.title("PDF of Undecided Prior Distribution")
plt.xlim(0,1)
```
